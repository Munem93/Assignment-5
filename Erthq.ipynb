```python
## Filtering the data to be more specific by covering one region with specific specifications 
new_df= df.loc[(df['place'].str.contains('CA')) & (df['type']=="earthquake") 
               & (df['magType']== "md") & (df['gap']<50) & (df['mag']<1)
               &(df['depth']>1)]
new_df.to_csv('filtered.csv')
new_df.reset_index(drop=True, inplace=True)
new_df





```
